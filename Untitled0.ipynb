The Breast Cancer Wisconsin dataset contains features computed from a digitized image of a fine needle aspirate (FNA) of a breast mass. They describe characteristics of the cell nuclei present in the image.

Features:

- radius (mean of distances from center to points on the perimeter): The average radius of the tumor mass.
- texture (standard deviation of gray-scale values): A measure of the tumor's texture.
- perimeter: The perimeter of the tumor mass.
- area: The area of the tumor mass.
- smoothness (local variation in radius lengths): A measure of the smoothness of the tumor.
- compactness (perimeter^2 / area - 1.0): A measure of how compact the tumor is.
- concavity (severity of concave portions of the contour): A measure of the severity of concave portions of the tumor.
- concave points (number of concave portions of the contour): The number of concave portions of the tumor.
- symmetry: A measure of the symmetry of the tumor.
- fractal dimension ("coastline approximation" - 1): A measure of the fractal dimension of the tumor.
Notes:

For each of the above features, the mean, standard error, and "worst" (largest) value were computed for each image, resulting in 30 features.

The variables related to tumor size and shape ("worst radius", "worst concave points", "worst perimeter", "worst area") stand out as the most influential in the model predictions, highlighting their crucial role in the classification between malignant and benign tumors.
